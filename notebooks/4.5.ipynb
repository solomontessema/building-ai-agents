```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

# Initialize the LLM

llm = ChatOpenAI(model="gpt-4", temperature=0.3)

# Define a prompt template
prompt = PromptTemplate(
input_variables=["task"],
template="""
You are an expert assistant.
Please complete the following task with clarity and precision:
{task}
Limit your response to 3 sentences.
"""
)

# Format and debug the prompt
formatted_prompt = prompt.format(task="Summarize the purpose of LangGraph in agent workflows")

# Run the LLM
response = llm.invoke(formatted_prompt)
print(response)


```
