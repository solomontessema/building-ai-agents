```python

from langgraph.graph import StateGraph, END
from typing import TypedDict

class GraphState(TypedDict):
    input: str
    trace: list
    output: str

def track(label):
    def inner(state):
        state.setdefault("trace", []).append(label)
        state["output"] = f"Reached {label}"
        return state
    return inner

graph = StateGraph(GraphState)
graph.add_node("step_a", track("A"))
graph.add_node("step_b", track("B"))

graph.set_entry_point("step_a")
graph.add_edge("step_a", "step_b")
graph.set_finish_point("step_b")

graph = graph.compile()

if __name__ == "__main__":
    result = graph.invoke({"input": "test"})
    print("Trace:", result["trace"])
    print("Output:", result["output"])

```
