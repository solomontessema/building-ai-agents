```python
from dotenv.main import load_dotenv
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

llm = ChatOpenAI(model="gpt-4", temperature=0)

# Step 1: Intent classification
intent_prompt = PromptTemplate.from_template(
"Classify the intent of this query: {query}"
)

# Step 2: Clarifying question generation
followup_prompt = PromptTemplate.from_template(
"Ask a clarifying question based on intent: {intent}"
)

# Chain the prompts using LCEL mapping
pipeline = {
"intent": intent_prompt
} | followup_prompt | llm | StrOutputParser()

# Example input
response = pipeline.invoke({"query": "Find hotels in Paris under $150"})
print(response)
```
