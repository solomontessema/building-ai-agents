```python
from langchain_core.prompts import FewShotPromptTemplate, PromptTemplate

# Define examples for few-shot prompting
examples = [
    {"input": "Hello", "output": "Bonjour"},
    {"input": "Thank you", "output": "Merci"},
]

# Template for formatting each example
example_prompt = PromptTemplate(
    input_variables=["input", "output"],
    template="Input: {input}\nOutput: {output}"
)

# Prompt template that includes the examples and the query
few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    suffix="Input: {user_input}\nOutput:",
    input_variables=["user_input"]
)

# Format the final prompt with a real user input
final_prompt = few_shot_prompt.format(user_input="Good morning")
print(final_prompt)
```
