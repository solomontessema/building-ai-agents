### Basic LangChain Runnable to answer a natural language question.


```python
!pip install -qU langchain==1.1.0 langchain-openai==1.1.0 python-dotenv==1.1.1
```

```python
# runnable_app.py
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

prompt = PromptTemplate(
    input_variables=["question"],
    template="Answer clearly and concisely: {question}"
)

# Build the pipeline: Prompt → LLM → Parser
runnable = prompt | llm | StrOutputParser()

response = runnable.invoke({"question": "What is LangChain used for?"})
print(response)
```
